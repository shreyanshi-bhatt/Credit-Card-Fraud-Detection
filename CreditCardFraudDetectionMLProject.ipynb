```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import os
print(os.listdir(path))
```

```python
import pandas as pd
df = pd.read_csv(f"{path}/creditcard.csv")
print(df.head())
```

```python
df.tail(3)
```

```python
df.shape
```

```python
print("No of Rows: ", df.shape[0])
print("No of Cols: ", df.shape[1])
```

```python
df.info()
```

```python
df.isnull().sum() # check no of null values in each col
```

```python
# Use scaling for Amount because rest all in same range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['Amount'] = sc.fit_transform(pd.DataFrame(df['Amount']))
df.head(3)
```

```python
df.drop(['Time'], axis=1, inplace=True)
```

```python
df.duplicated().any()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['Class'].value_counts() # to check if there is imbalance in the target variable
```

```python
# To handle imbalanced Dataset, use Undersampling or Oversampling
# Recall & f1-score will be low for imbalanced Datasets
# Acuracy might decrease after undersampling/oversampling

# Undersampling
normal = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]
print("Normal Transactions: ", normal.shape)
print("Fraudulent Transactions: ", fraud.shape)
```

```python
normal_sample = normal.sample(n=473) # undersampling
new_data = pd.concat([normal_sample, fraud], axis=0, ignore_index = True)
new_data['Class'].value_counts()
```

```python
X = new_data.drop('Class', axis=1)
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=27)
```

```python
# Try different ML Algos
# Logistic Reg
from sklearn.linear_model import LogisticRegression
LRModel = LogisticRegression()
LRModel.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = LRModel.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
DTModel = DecisionTreeClassifier()
DTModel.fit(X_train, y_train)
```

```python
y_pred2 = DTModel.predict(X_test)
print(accuracy_score(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RFModel = RandomForestClassifier()
RFModel.fit(X_train, y_train)
y_pred3 = RFModel.predict(X_test)
print("Accuracy: ",accuracy_score(y_test, y_pred3))
print("Confusion Matrix: ",confusion_matrix(y_test, y_pred3))
print("Classification Report: ",classification_report(y_test, y_pred3))
```

```python
# Visualize perf of all 3 models
comparision = pd.DataFrame({'Models':['LR', 'DT', 'RF'],
              'Accuracy':[accuracy_score(y_test, y_pred)*100,
                          accuracy_score(y_test, y_pred2)*100,
                          accuracy_score(y_test, y_pred3)*100]})
```

```python
comparision
```

```python
import seaborn as sns
```

```python
sns.barplot(x=comparision['Models'], y=comparision['Accuracy'])
```

```python
# Oversampling = Data Augmentation
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)
y_res.value_counts()
```

```python
# Try different ML Algos
# Logistic Reg
from sklearn.linear_model import LogisticRegression
LRModel = LogisticRegression()
LRModel.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = LRModel.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
DTModel = DecisionTreeClassifier()
DTModel.fit(X_train, y_train)
```

```python
y_pred2 = DTModel.predict(X_test)
print(accuracy_score(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RFModel = RandomForestClassifier()
RFModel.fit(X_train, y_train)
y_pred3 = RFModel.predict(X_test)
print("Accuracy: ",accuracy_score(y_test, y_pred3))
print("Confusion Matrix: ",confusion_matrix(y_test, y_pred3))
print("Classification Report: ",classification_report(y_test, y_pred3))
```
